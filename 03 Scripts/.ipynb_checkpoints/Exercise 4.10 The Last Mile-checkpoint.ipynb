# Creating an exclusion flag for low-activity customers

```python
# Create an exclusion flag for low-activity customers
df['exclusion_flag'] = df.groupby('user_id')['order_id'].transform('count').apply(lambda x: 'Exclude' if x < 5 else 'Include')

# Filter the data to exclude low-activity customers
df_filtered = df[df['exclusion_flag'] == 'Include']

# Count unique users for each flag
unique_user_count = df['exclusion_flag'].value_counts()

# Print the unique user count
print("Unique User Count by Exclusion Flag:")
print(unique_user_count)
```

```python
# Export the filtered data as a pickle file
df_filtered.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'filtered_data.pkl'))
```

# Exporting the data of the excluded customers

```python
# Filter the data for excluded customers
df_excluded = df[df['exclusion_flag'] == 'Exclude']

# Export the excluded customers' data to an Excel file
df_excluded.to_excel(os.path.join(path, '02 Data', 'Prepared Data', 'excluded_customers.xlsx'), index=False)
```

# Creating a profiling variable based on age and number of dependents

```python
# Create a copy of the dataframe to avoid modifying the original
df_copy = df_filtered.copy()

# Define a function to assign profiles based on age and number of dependents
def assign_customer_profile(row):
    if row['age'] < 30 and row['number_of_dependents'] == 0:
        return 'Single adult'
    elif row['age'] < 40 and row['number_of_dependents'] > 0:
        return 'Young parent'
    elif 40 <= row['age'] < 65:
        return 'Middle-aged'
    elif row['age'] >= 65:
        return 'Old and retired'
    else:
        return 'Other'
```

```python
# Apply the function to create a new column for customer profiles in the copy
df_copy['customer_age_depn'] = df_copy.apply(assign_customer_profile, axis=1)
```

```python
# Group the data to include dependents in the profiles for visualization later
profile_dependents = df_copy.groupby(['customer_age_depn', 'number_of_dependents']).size().reset_index(name='count')
```

```python
# Check the distribution of profiles
print(df_copy['customer_age_depn'].value_counts())
```

# Pie Chart : based on age and number of dependents

```python
# Get the distribution of customer profiles
profile_distribution = df_copy['customer_age_depn'].value_counts()

# Define a new set of colors for the pie chart
colors = ['#FF9999', '#66B2FF', '#FFCC99', '#99FF99', '#D699FF']  # Soft pastel colors

# Create the pie chart
plt.figure(figsize=(8, 8))
pie1 = profile_distribution.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=colors, 
    explode=(0.11, 0, 0, 0, 0),  # Highlight the first segment slightly
    shadow=True
)

# Add a title
plt.title('Distribution of Customer Profiles', fontsize=16)

# Remove the y-axis label for better visualization
plt.ylabel('')

# Show the pie chart
plt.show()
```

```python
# Save the pie chart to the specified path
pie1.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Customer_profile_Age_Dependents.png'))
```

# Summary of how many times each department_id appears for each customer_profile

```python
# Crosstab between customer_profile and department_id
crosstab_dept = pd.crosstab(df_copy['customer_age_depn'], df_copy['department_id'], normalize='index')

# Display the crosstab
print(crosstab_dept)

# Export the crosstab as a pickle file
crosstab_dept.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'crosstab_customer_profile_department.pkl'))
```

# Bar Chart : compare the distributions of each department across different customer profiles

```python
# Plot the bar chart
Bar_c = crosstab_dept.plot(kind='bar', figsize=(12, 8), cmap='viridis', width=0.8)

# Add chart labels and title
plt.title('Customer Profiles vs Department ID', fontsize=16)
plt.xlabel('Customer Profiles', fontsize=12)
plt.ylabel('Proportion of Orders', fontsize=12)

# Modify x-axis labels to be horizontal
Bar_c.set_xticklabels(['Middle-aged', 'Old and Retired', 'Other', 'Single-adult', 'Young Parent'], rotation=0, ha='center')

# Add a legend with adjusted position
plt.legend(title='Department ID', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the chart
plt.show()
```

```python
# Save the chart to the specified path
Bar_c.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Customer_profile_crosstab_dept_id.png'))
```

# Aggregate the max, mean and min variables on a customer-profile level for usage frequency(max_order) and expenditure(prices)
