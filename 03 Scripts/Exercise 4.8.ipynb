 Analysis of Results: Full Dataset vs Subset

 Key Observations:
 
1. **Mean Values:**
   - The mean `order_number` values for most `department_id`s in the full dataset (`ords_prods_merge_copy`) are **slightly higher** compared to those in the subset (`df`).

     - This suggests that the subset may underrepresent users with higher order numbers, leading to slightly lower averages.

2. **Minimum and Maximum Values:**
   - The minimum `order_number` value for all departments is **1** in both datasets, indicating consistency.
   - The maximum `order_number` value is **99** for nearly all departments in both datasets, except for `department_id 21`, where:
     - Subset maximum = **97**
     - Full dataset maximum = **99**

3. **Variability Across Departments:**
   - Departments with higher mean values, such as `department_id 10` and `department_id 21`, show slight increases in the full dataset compared to the subset.
   - Departments with lower mean values, such as `department_id 1` and `department_id 5`, also exhibit increases but to a smaller extent.

4. **Subset Representativeness:**
   - The subset (`df`) appears to be a reasonable representation of the overall dataset in terms of distribution. However, the subset may slightly **underrepresent users with higher order numbers** in some departments, leading to marginally lower means.

  Conclusion:
  
- The analysis shows that while the subset (`df`) provides a similar overall pattern, it **slightly underestimates mean order numbers** for certain departments compared to the full dataset (`ords_prods_merge_copy`).
- For the most accurate insights, it is recommended to use the **full dataset** to ensure all patterns and variations are captured.


```python
### Create a flag using the following criteria
# If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
# If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
# If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

# Create a new column 'max_order' with the maximum 'order_number' for each 'user_id'
ords_prods_merge_copy['max_order'] = ords_prods_merge_copy.groupby(['user_id'])['order_number'].transform('max')

# Confirm the column is created successfully
print("Column 'max_order' created successfully.")
```

```python
ords_prods_merge_copy['max_order'].value_counts(dropna = False)
```

```python
ords_prods_merge_copy.head(9)
```

```python
# Create a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value
ords_prods_merge_copy.loc[ords_prods_merge_copy['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge_copy.loc[(ords_prods_merge_copy['max_order'] <= 40) & (ords_prods_merge_copy['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge_copy.loc[ords_prods_merge_copy['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Print the frequency of the new “loyalty_flag” column
ords_prods_merge_copy.groupby(['loyalty_flag']).user_id.nunique()
```

```python
ords_prods_merge_copy[['user_id', 'loyalty_flag', 'order_number']].head(6)
```

```python
# Group the data by 'loyalty_flag' and calculate basic statistics for the 'prices' column
loyalty_price_stats = ords_prods_merge_copy.groupby('loyalty_flag')['prices'].describe()

# Calculate the unique user count for each loyalty flag
unique_user_count = ords_prods_merge_copy.groupby('loyalty_flag')['user_id'].nunique()

# Combine the statistics with the unique user count
loyalty_price_stats['unique_user_count'] = unique_user_count

# Display the updated statistics
print(loyalty_price_stats)
```

```python
# Calculate the mean price for each user and assign a spending flag
ords_prods_merge_copy['spending_flag'] = ords_prods_merge_copy.groupby('user_id')['prices'].transform('mean').apply(
    lambda x: 'Low spender' if x < 10 else 'High spender'
)

# Calculate the unique user count for each spending flag
unique_user_count = ords_prods_merge_copy.groupby('spending_flag')['user_id'].nunique()

# Display the unique user count
print("Unique User Count by Spending Flag:")
print(unique_user_count)
```

```python
# Create the 'order_frequency_flag' column
ords_prods_merge_copy['order_frequency_flag'] = ords_prods_merge_copy.groupby('user_id')['days_since_prior_order'].transform('median').apply(
    lambda x: 'Non-frequent customer' if x > 20 else 
              'Regular customer' if 10 < x <= 20 else 
              'Frequent customer'
)

# Calculate unique user IDs in each group
unique_user_count = ords_prods_merge_copy.groupby('order_frequency_flag')['user_id'].nunique()

# Display the unique user count
print("Unique User Count by Order Frequency Flag:")
print(unique_user_count)
```

```python
# Downcasting numerical columns to reduce memory usage in ords_prods_merge_copy
for column in ords_prods_merge_copy.columns:
    if ords_prods_merge_copy[column].dtype == 'float64':
        ords_prods_merge_copy[column] = pd.to_numeric(ords_prods_merge_copy[column], downcast='float')
    elif ords_prods_merge_copy[column].dtype == 'int64':
        ords_prods_merge_copy[column] = pd.to_numeric(ords_prods_merge_copy[column], downcast='integer')

print("Downcasting completed for ords_prods_merge_copy.")
```
